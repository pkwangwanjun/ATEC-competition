```python
# -*- coding:utf-8 -*-
%cd E:\竞赛\atec
import numpy as np
import pandas as pd
```

```python
NUM_DATASET = 10
```

```python
data = pd.read_csv("atec_anti_fraud_train.csv")
```

```python
import lightgbm as lgb
```

```python
data["label"] = data.label.replace(-1,1)
data["weight"] = data.label.apply(lambda x:100 if x==1 else 1)
```

```python
one = data[data.label==1]
zero = data[data.label==0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
one_train,one_val = train_test_split(one,test_size=1000,stratify=one.date,random_state=0)
```

```python
train_size = one_train.shape[0]
```

```python
zeroset = []
for i in range(NUM_DATASET):
    _,zero_tmp = train_test_split(zero,test_size=train_size,stratify=zero.date,random_state=i)
    zeroset.append(zero_tmp)
```

```python
dataset = []
for i in range(NUM_DATASET):
    dataset_tmp = pd.concat([zeroset[i],one],axis=0)
    dataset.append(dataset_tmp)
```

```python
_,zero_val= train_test_split(zero,test_size=1000*100,stratify=zero.date,random_state=42)
val = pd.concat([zero_val,one_val],axis=0)
```

```python
params = {
    "task": "train",
    "boosting_type": "gbdt",
    "objective": "binary",
    "metric": "auc",
    "learning_rate": 0.01,
    "num_leaves": 255,
    "max_depth": 10,
    "min_child_samples": 200,
    "bagging_fraction": 0.95,
    "bagging_freq": 1,
    "feature_fraction": 0.5,
}
#"subsample_for_bin": 1000000,
```

```python
lgbmodellist = []
for i in range(NUM_DATASET):
    data = dataset[i]
    d_train = lgb.Dataset(data[data.columns.difference(["id","label","weight","date"])],label=data.label,weight=data.weight)
    d_val = lgb.Dataset(val[val.columns.difference(["id","label","weight","date"])],label=val.label,reference=d_train)
    bst = lgb.train(params,d_train,200,early_stopping_rounds=10,valid_sets=[d_train,d_val])#
    lgbmodellist.append(bst)
```

# Validation

```python
from sklearn.metrics import roc_curve
def score_atc(y_true,y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_true,y_pred_prob, pos_label=1)
    score=0.4*tpr[(fpr>=0.001) & (fpr<=0.0011)][0]+0.3*tpr[(fpr>=0.005) & (fpr<=0.0051)][0]+0.3*tpr[(fpr>=0.01) & (fpr<=0.011)][0]
    return score
```

```python
res = []
for i in range(NUM_DATASET):
    model = lgbmodellist[i]
    pred = model.predict(val[val.columns.difference(["id","label","weight","date"])],num_iteration = model.best_iteration)
    res.append(pred)
result = np.array(res)
result = result.mean(axis=0)
```

```python
print(params)
print(score_atc(val.label,result))
```
